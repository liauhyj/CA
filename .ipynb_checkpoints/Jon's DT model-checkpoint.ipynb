## Reading the file and encoding of columns

```python
import pandas as pd

df = pd.read_csv("mushrooms_1000rows.csv")

df
```

```python
class_dict = {"p": 0,
              "e": 1}

df["class"] = df["class"].map(class_dict)

df[["class"]].head()
```

```python
# "b", "c", "f", "k", "s", "x"

cap_shape_list =  ["b", "c", "f", "k", "s", "x"]

cap_shape_dict = {}

for i in range(len(cap_shape_list)):
    cap_shape_dict[cap_shape_list[i]] = i

df["cap-shape"] = df["cap-shape"].map(cap_shape_dict)

df[["cap-shape"]].head()
```

```python
# "f", "g", "s" and "y"

cap_surface_list =  ["f", "g", "s", "y"]

cap_surface_dict = {}

for i in range(len(cap_surface_list)):
    cap_surface_dict[cap_surface_list[i]] = i

df["cap-surface"] = df["cap-surface"].map(cap_surface_dict)

df[["cap-surface"]].head()
```

```python
# "b", "c", "e", "g", "n", "p", "r", "u", "w" and "y"

cap_color_list = ["b", "c", "e", "g", "n", "p", "r", "u", "w", "y"]

cap_color_dict = {}

for i in range(len(cap_color_list)):
    cap_color_dict[cap_color_list[i]] = i

df["cap-color"] = df["cap-color"].map(cap_color_dict)

df[["cap-color"]].head()
```

```python
bruises_dict = {"f":0,
                "t": 1}

df["bruises"] = df["bruises"].map(bruises_dict)
df[["bruises"]].head()
```

```python
# "a", "c", "f", "l", "m", "n", "p", "s" and "y"

odor_list = ["a", "c", "f", "l", "m", "n", "p", "s", "y"]
odor_dict = {}

for i in range(len(odor_list)):
    odor_dict[odor_list[i]] = i
    
df["odor"] = df["odor"].map(odor_dict)
df[["odor"]].head()
```

```python
# "a", "c", "n", "s", "v" and "y"

population_list = ["a", "c", "n", "s", "v","y"]
population_dict = {}

for i in range(len(population_list)):
    population_dict[population_list[i]] = i

df["population"] = df["population"].map(population_dict)
df[["population"]].head()
```

```python
# "d", "g", "l", "m", "p", "u" and "w"

habitat_list = ["d", "g", "l", "m", "p", "u", "w"]
habitat_dict = {}

for i in range(len(habitat_list)):
    habitat_dict[habitat_list[i]] = i
                
df["habitat"] = df["habitat"].map(habitat_dict)
df[["habitat"]].head()
```

```python
# test if encoding has been done correctly

df.dropna(inplace = True)

df
```

## Training of DT model

```python
# splitting the data in training and test datasets

from sklearn.model_selection import train_test_split

y = df["class"]
x = df.iloc[:, 1: df.shape[1]]

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0)

x_train.head()
```

```python
y_train.head()
```

```python
# train the model

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

## Validation and prediction of DT model

```python
# validate the model using the test dataset

y_pred = dt.predict(x_test)

print(y_test)
print(y_pred)
```

```python
# predict the class when cap-shape = 2, cap-surface = 1, cap-color = 5, bruises = 1,
# odor = 6, population = 3, habitat = 4

print(dt.predict([[2, 1, 5, 1, 6, 3, 4]]))
```

```python
# get the accuracy score of the dt model

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
# confusion matrix of the model

from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

## Visualising the DT model

```python
from sklearn import tree
import graphviz
from graphviz import Source

# display the tree, with no output file

Source(tree.export_graphviz(dt, out_file=None, class_names=["p", "e"], 
                            feature_names= x_train.columns))
```

## Feature Engineering of the previous DT model

```python
# feature engineering can delete habitat and cap-shape from x_train

del x_train["habitat"]
del x_train["cap-shape"]
x_train.head()
```

```python
# train the model again using new x_train

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

## Validation and prediction of new DT model

```python
# validate the model using the test dataset

del x_test["habitat"]
del x_test["cap-shape"]

y_pred = dt.predict(x_test)

print(y_test)
print(y_pred)
```

```python
# predict the class when cap-shape = 2, cap-surface = 1, cap-color = 5, bruises = 1,
# odor = 6, population = 3

print(dt.predict([[1, 5, 1, 6, 3]]))
```

```python
# get the accuracy score of the new dt model

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
# confusion matrix of the model

from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

## Visualising the new DT model

```python
from sklearn import tree
import graphviz
from graphviz import Source

# display the tree, with no output file

Source(tree.export_graphviz(dt, out_file=None, class_names=["p", "e"], 
                            feature_names= x_train.columns))
```
