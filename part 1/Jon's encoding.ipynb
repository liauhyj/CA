```python
import pandas as pd

df = pd.read_csv("mushrooms_1000rows.csv")

df
```

```python
class_dict = {"p": 0,
              "e": 1}

df["class"] = df["class"].map(class_dict)

df[["class"]].head()
```

```python
# "b", "c", "f", "k", "s", "x"

cap_shape_list =  ["b", "c", "f", "k", "s", "x"]

cap_shape_dict = {}

for i in range(len(cap_shape_list)):
    cap_shape_dict[cap_shape_list[i]] = i

df["cap-shape"] = df["cap-shape"].map(cap_shape_dict)

df[["cap-shape"]].head()
```

```python
# "f", "g", "s" and "y"

cap_surface_list =  ["f", "g", "s", "y"]

cap_surface_dict = {}

for i in range(len(cap_surface_list)):
    cap_surface_dict[cap_surface_list[i]] = i

df["cap-surface"] = df["cap-surface"].map(cap_surface_dict)

df[["cap-surface"]].head()
```

```python
# "b", "c", "e", "g", "n", "p", "r", "u", "w" and "y"

cap_color_list = ["b", "c", "e", "g", "n", "p", "r", "u", "w", "y"]

cap_color_dict = {}

for i in range(len(cap_color_list)):
    cap_color_dict[cap_color_list[i]] = i

df["cap-color"] = df["cap-color"].map(cap_color_dict)

df[["cap-color"]].head()
```

```python
bruises_dict = {"f":0,
                "t": 1}

df["bruises"] = df["bruises"].map(bruises_dict)
df[["bruises"]].head()
```

```python
# "a", "c", "f", "l", "m", "n", "p", "s" and "y"

odor_list = ["a", "c", "f", "l", "m", "n", "p", "s", "y"]
odor_dict = {}

for i in range(len(odor_list)):
    odor_dict[odor_list[i]] = i
    
df["odor"] = df["odor"].map(odor_dict)
df[["odor"]].head()
```

```python
# "a", "c", "n", "s", "v" and "y"

population_list = ["a", "c", "n", "s", "v","y"]
population_dict = {}

for i in range(len(population_list)):
    population_dict[population_list[i]] = i

df["population"] = df["population"].map(population_dict)
df[["population"]].head()
```

```python
# "d", "g", "l", "m", "p", "u" and "w"

habitat_list = ["d", "g", "l", "m", "p", "u", "w"]
habitat_dict = {}

for i in range(len(habitat_list)):
    habitat_dict[habitat_list[i]] = i
                
df["habitat"] = df["habitat"].map(habitat_dict)
df[["habitat"]].head()
```

```python
# test if encoding has been done correctly

df.dropna(inplace = True)

df
```
